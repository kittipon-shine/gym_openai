# MountainCar

- gym: https://gym.openai.com/envs/MountainCar-v0/
- wiki: https://github.com/openai/gym/wiki/MountainCar-v0

```python
import gym
```

```python
env_name = 'MountainCar-v0'
env = gym.make(env_name)
```

```python
class Agent():
    def __init__(self, env):
        self.action_size = env.action_space.n
        print("action_size:", self.action_size)
        
    def get_action(self, state):
        velocity = state[1]

        if (velocity < 0):
            action = 0 #push left
        else:
            action = 2 #push right
        return action
    """
    velocity เป็นลบคือรถไปทางซ้าย เราจะดันรถไปทางซ้ายเพื่อขึ้นเนินฝั่งซ้าย เพื่อให้ตอนไหลลงมาจะได้มีแรงส่งไปทางเนินขวา

    velocity เป็นบวกคือรถไปทางขวา เราจะดันรถไปทางขวาเพื่อพลักขึ้นเนินขวาจะได้ขึ้นไปหาธงได้
    """
```

```python
agent = Agent(env)
state = env.reset()
# state คือ observation

for i in range(200):
    print('state',i,'velocity:', state[1])

#     action = env.action_space.sample()

    action = agent.get_action(state)
    state, reward, done, info = env.step(action)
    env.render()
```
