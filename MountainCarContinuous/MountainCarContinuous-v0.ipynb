```python
import gym
```

```python
env_name = "MountainCarContinuous-v0"
env = gym.make(env_name)
```

```python
class Agent():
    def __init__(self, env):
        self.action_size = env.observation_space.shape[0]
        print(self.action_size)

    def get_action(self, state):
        velocity = state[1]
        
        if velocity < 0:
            action = 0
        else:
            action = 1
        return [action] #ค่า action ที่เป็น sample เป็น list
```

```python
agent = Agent(env)
state = env.reset()
# state คือ observation

for i in range(200):
    print('state',i,'velocity:', state[1])

#     action = env.action_space.sample()
#     print(action)
    action = agent.get_action(state)
    state, reward, done, info = env.step(action)
    env.render()
    
print(reward)
```
