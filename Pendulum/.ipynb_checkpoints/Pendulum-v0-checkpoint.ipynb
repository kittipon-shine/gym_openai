# Pendulum

- gym: https://gym.openai.com/envs/Pendulum-v0/
- wiki: https://github.com/openai/gym/wiki/Pendulum-v0

```python
import gym
```

```python
env_name = 'Pendulum-v0'
env = gym.make(env_name)
```

```python
class Agent():
    def __init__(self, env):
        self.action_size = env.observation_space.shape[0]
        print('action_size:', self.action_size)
    def get_action(self, state):
        cos = state[0]
        sin = state[1]
        
        if (cos <= 0.8) and (sin >= 0): #ล่างซ้าย
            action = -10
        elif (cos <= 0.8) and (sin <= 0): #ล่างขวา
            action = 10
        elif (cos > 0.8) and (sin > 0): #บนซ้าย
            action = -5
        elif (cos > 0.8) and (sin < 0): #บนขวา
            action = 5
        else:
            action = 0
        return [action]
```

- action: +right, -left

- cos: +up, -down
- sin: +left, -right
- theta dot: 

```python
agent = Agent(env)
state = env.reset()

for i in range(1000):
    print('state',i,'velocity:', state[0], state[1])
    
#     action = env.action_space.sample()
    action = agent.get_action(state)
    print('action:', action)

    state, reward, done, info = env.step(action)
    env.render()
print('reward:', reward)
```
